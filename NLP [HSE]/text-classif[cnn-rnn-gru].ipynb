## Удалим стоп слова

```python
import nltk
from nltk.corpus import stopwords
from collections import Counter
import string

# Загрузка стоп-слов для английского
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
```

```python
print(f'symbols that will be removed from text {string.punctuation}')
translator = str.maketrans('', '', string.punctuation) # создаем список знаков препинания которые должны быть удалены
# первые два аргумента пустые строки потому что нам не нужно заменять одни символы на другие, а надо почистить текст (удалять)

texts_without_sw = [[] for _ in range(len(list_of_texts))] # создаем список списков куда будем сохранять обработанные тексты
for i in range(len(list_of_texts)):
    temp = list_of_texts[i][0].translate(translator).lower().split() # удаляем все знаки, переводим в лоуер кейс и сплитим
    #print(temp)
    #break 
    texts_without_sw[i] = [word for word in temp if word not in stop_words] # отбираем все кроме стоп слов из библиотеки nltk
```

```python
# костыльно все получается конечно ахахаха) но теперь надо объединить все списки в единый спискок и решить поставленную задачу
from itertools import chain
flat_list = list(chain.from_iterable(texts_without_sw)) 
```

```python
# считаем частоту слов
word_counts = Counter(flat_list)

# находим топ кэс самых частых слов
top_1000 = word_counts.most_common(1000)

print(top_1000)
```

# Часть 2. [2 балла] Модели представления слов

```python
# Все сразу не удается загрузить, не хватает опертивной памяти, так что будем загружать по файлу отдельно
# и сразу препрецессить текст: убираем все знаки препинания + переводим в нижний регистр и преобразуем в список
# в конце избавляемся от списка списков и сохраняем все в единый файл
import string
import os
from itertools import chain

path = '/kaggle/input/nlp-data/fanfiction_texts/fanfiction_texts/'
files_path = os.listdir('/kaggle/input/nlp-data/fanfiction_texts/fanfiction_texts') # список файлов
out_path = '/kaggle/working/'

translator = str.maketrans('', '', string.punctuation) # транслятор для того что бы почистить текст от знаков препинания
```

```python
from tqdm.notebook import tqdm
# создаем файл
with open(out_path + 'giga_file.txt', 'w') as file:
    pass

# функция для записи в единый файл всех предобработынных текстов данных
def fill_file(in_path, out_path):
    x = []
    with open(out_path, 'a') as out:
        with open(in_path, 'r') as file:
            for line in file:
                x = line.strip().translate(translator).lower().split()
                x.append('\n') # как я понял надо добавлять символ перевода каретки иначе потом итоговый файл читается весь как единая строка
                # что бы избежать это и читать его кусками я добавил \n
                x = ' '.join(x) # получаем единую строку
                out.write(x) # записываем ее в файл добавляя в конец

for curr_path in tqdm(files_path):
    fill_file(path + curr_path, out_path + 'giga_file.txt')
```

```python
# генератор, который будет выдавать готовый текст для обучения
class MyGenerator():
    def __init__(self, path):
        self.path = path
    
    def __iter__(self):
        with open(self.path, 'r') as file:
            for line in file:
                yield line.strip().split()
```

```python
from gensim.models import Word2Vec
corpus = MyGenerator('/kaggle/working/giga_file.txt')
wv_model = Word2Vec(sentences=corpus, vector_size=64, window=5, min_count=1, workers=4)
```

```python
wv_model.save("/kaggle/working/word2vec.model")
```

```python
UNK_IX, PAD_IX = map(token_to_id.get, [UNK, PAD])
SENT_LEN = int(median) # будем использовать медианную длину в качестве максимальной длины предложений

def as_matrix(sequences, max_len=SENT_LEN):
    """ Convert a list of tokens into a matrix with padding """
    if isinstance(sequences[0], str):
        sequences = list(map(str.split, sequences))

    #max_len = min(max(map(len, sequences)), max_len or float('inf'))

    matrix = np.full((len(sequences), max_len), np.int32(PAD_IX))
    for i,seq in enumerate(sequences):
        row_ix = [token_to_id.get(word, UNK_IX) for word in seq[:max_len]]
        matrix[i, :len(row_ix)] = row_ix

    return matrix
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from keras.utils import to_categorical

device = 'cuda' if torch.cuda.is_available() else 'cpu'

def to_tensors(batch, device):
    batch_tensors = dict()
    for key, arr in batch.items():
        if key == "Text":
            batch_tensors[key] = torch.tensor(arr, device=device, dtype=torch.int64)
        else:
            batch_tensors[key] = torch.tensor(arr, device=device)
    return batch_tensors


def make_batch(data, max_len=128, word_dropout=0, device=device):
    """
    Creates a keras-friendly dict from the batch data.
    :param word_dropout: replaces token index with UNK_IX with this probability
    :returns: a dict with {'title' : int64[batch, title_max_len]
    """
    batch = {}
    batch["Text"] = as_matrix(data['text'].values, max_len)
    
    if word_dropout != 0:
        batch["Text"] = apply_word_dropout(batch["Text"], 1. - word_dropout)

    if 'label' in data.columns:
        batch['label'] = to_categorical(data['label'], num_classes=85) 

    return to_tensors(batch, device)

def apply_word_dropout(matrix, keep_prop, replace_with=UNK_IX, pad_ix=PAD_IX,):
    dropout_mask = np.random.choice(2, np.shape(matrix), p=[keep_prop, 1 - keep_prop])
    dropout_mask &= matrix != pad_ix
    return np.choose(dropout_mask, [matrix, np.full_like(matrix, replace_with)])
```

```python
import matplotlib.pyplot as plt

def plot_training_results(results, title=None):
    """
    Функция для отрисовки результатов обучения с двумя цветами: розовый (train) и черный (val).

    Аргумент:
    results -- словарь, содержащий результаты по обучению и валидации:
               {'train': {'loss': [...], 'f1': [...]},
                'val': {'loss': [...], 'f1': [...]}}

    """
    epochs = range(1, len(results['train']['loss']) + 1)

    # Создаем подграфики для loss и f1-score
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

    # Цвета
    train_color = 'red'  # Розовый
    val_color = 'green'   # Черный

    # Отрисовка потерь
    ax1.plot(epochs, results['train']['loss'], label='Train Loss', marker='o', color=train_color)
    ax1.plot(epochs, results['val']['loss'], label='Validation Loss', marker='o', color=val_color)
    ax1.set_title('Loss per Epoch')
    ax1.set_xlabel('Epochs')
    ax1.set_ylabel('Loss')
    ax1.legend()
    ax1.grid(True)

    # Отрисовка F1-score
    ax2.plot(epochs, results['train']['f1'], label='Train F1', marker='o', color=train_color)
    ax2.plot(epochs, results['val']['f1'], label='Validation F1', marker='o', color=val_color)
    ax2.set_title('F1-Score per Epoch')
    ax2.set_xlabel('Epochs')
    ax2.set_ylabel('F1-Score')
    ax2.legend()
    ax2.grid(True)
    if title:
        plt.suptitle(title, fontsize=16)
    # Отображаем графики
    plt.tight_layout()
    plt.show()
```

```python
from sklearn.metrics import f1_score
TARGET_COLUMN = 'label'

def iterate_minibatches(data, batch_size=256, shuffle=True, cycle=False, device=device):
    """ iterates minibatches of data in random order """
    while True:
        indices = np.arange(len(data))
        if shuffle:
            indices = np.random.permutation(indices)

        for start in range(0, len(indices), batch_size):
            batch = make_batch(data.iloc[indices[start : start + batch_size]], device=device)
            yield batch

        if not cycle: break 
            
def calculate_accuracy(output, target):
    _, predicted = torch.max(output, dim=1)
    _, target= torch.max(target, dim=1)
    correct = (predicted == target).sum().item()
    total = target.size(0)
    accuracy = correct / total
    return accuracy
 

def print_metrics(model, data, batch_size=16, device=device):
    model.eval()
    with torch.no_grad():
        acc = 0
        ctr = 0 
        val_preds = []
        val_true = []
        
        for batch in iterate_minibatches(data, batch_size=batch_size, shuffle=False, device=device):
            batch_pred = model(batch)
            
            true_labels = torch.argmax(batch[TARGET_COLUMN], dim=1).cpu().numpy()
            pred_labels = torch.argmax(batch_pred, dim=1).cpu().numpy()
            
            val_preds.extend(pred_labels)
            val_true.extend(true_labels)
            
            acc += calculate_accuracy(batch_pred, batch[TARGET_COLUMN])
            ctr += 1
    acc /= ctr
    f1_macro = f1_score(val_true, val_preds, average='macro')
    f1_w = f1_score(val_true, val_preds, average='weighted')
    f1_micro = f1_score(val_true, val_preds, average='micro')
    print("\nTest (Validation) Results:")
    print("Accuracy: %.5f" % acc)
    print("Macro F1: %.5f" % f1_macro)
    print("Weighted F1: %.5f" % f1_w)
    print("Micro F1: %.5f" % f1_micro)
    return f1_macro
```

Трейн луп писал сам! в него я добавил и сохранение лучшего веса по результатам f1-macro, и шедулер добавил

```python
from tqdm.notebook import tqdm
import copy
def train_loop(model,
               data_train,
               data_val,
               loss_fn, 
               opt,  
               batch_size, 
               epochs,
               device,
               save_best_weights=True,
               print_each = 5,
               scheduler=None):
    

    best_weights = None
    best_f1 = -float('inf')
    acc = loss_score = ctr = 0
    tr_preds = []
    tr_true = []
    metrics = {'train' : {'loss' : [], 'f1': []},
              'val': {'loss' : [], 'f1': []}}
    for epoch in range(epochs):
        print(f'epoch: {epoch+1}')
        model.train()
        for i, batch in tqdm(enumerate(
            iterate_minibatches(data_train, batch_size=batch_size, device=device)),
            total=len(data_train) // batch_size
        ):
            y_pred = model(batch)
            loss = loss_fn(y_pred, batch['label'])
            opt.zero_grad()
            loss.backward()
            opt.step()
            
            # считаем и сохраняем метрики
            ctr += 1
            loss_score += loss.item()
            true_labels = torch.argmax(batch['label'], dim=1).cpu().numpy()
            pred_labels = torch.argmax(y_pred, dim=1).cpu().numpy()
            tr_preds.extend(pred_labels)
            tr_true.extend(true_labels)
            acc += calculate_accuracy(y_pred, batch['label'])
        if scheduler:
            scheduler.step()
        acc /= ctr
        loss_score /= ctr
        f1 = f1_score(tr_true, tr_preds, average='macro')
        if (epoch+1) % print_each == 0:
            print("\nTrain Results:")
            if scheduler:
                print(f'Learning rate: {scheduler.get_last_lr()}')
            print("Accuracy: %.5f" % acc)
            print("Macro F1: %.5f" % f1)
            
            
        metrics['train']['f1'].append(f1)
        metrics['train']['loss'].append(loss_score)
        acc = loss_score = ctr = 0
        tr_true = []
        tr_preds = []
  
        
        model.eval()
        with torch.no_grad():
            val_acc = val_ctr = val_loss = 0
            val_preds = []
            val_true = []
            
            for batch in iterate_minibatches(data_val, batch_size=batch_size, shuffle=False, device=device):
                y_pred = model(batch)
                loss = loss_fn(y_pred, batch['label'])
                
                true_labels = torch.argmax(batch['label'], dim=1).cpu().numpy()
                pred_labels = torch.argmax(y_pred, dim=1).cpu().numpy()
                val_preds.extend(pred_labels)
                val_true.extend(true_labels)
                val_acc += calculate_accuracy(y_pred, batch['label'])
                val_loss += loss.item()
                val_ctr += 1
        val_acc /= val_ctr 
        val_loss /= val_ctr
        val_f1 = f1_score(val_true, val_preds, average='macro')
        if (epoch+1) % print_each == 0:
            print("\nValid Results:")
            print("Accuracy: %.5f" % val_acc)
            print("Macro F1: %.5f" % val_f1)
        metrics['val']['f1'].append(val_f1)
        metrics['val']['loss'].append(val_loss)
        
        if val_f1 > best_f1:
            best_f1 = val_f1
            best_weights = copy.deepcopy(model.state_dict())
    
    if save_best_weights:
        model.load_state_dict(best_weights)
    return metrics
```

```python
class BidirRNNLayer(nn.Module):
    def __init__(self, emb_size, hid_size=64):
        super().__init__()
        
        self.hid_size = hid_size
        
        self.w_x_forward = nn.Parameter(torch.Tensor(emb_size, hid_size))
        self.w_h_forward = nn.Parameter(torch.Tensor(hid_size, hid_size))
        self.b_h_forward = nn.Parameter(torch.zeros(1, hid_size))
        
        self.w_x_backward = nn.Parameter(torch.Tensor(emb_size, hid_size))
        self.w_h_backward = nn.Parameter(torch.Tensor(hid_size, hid_size))
        self.b_h_backward = nn.Parameter(torch.zeros(1, hid_size))
        
        nn.init.xavier_uniform_(self.w_x_forward)
        nn.init.xavier_uniform_(self.w_h_forward)
        nn.init.xavier_uniform_(self.w_x_backward)
        nn.init.xavier_uniform_(self.w_h_backward)
        
        self.layer_norm_forward = nn.LayerNorm(hid_size)
        self.layer_norm_backward = nn.LayerNorm(hid_size)
        
        self.y = nn.Linear(hid_size * 2, hid_size * 2)
    
    def forward(self, x):
        batch_size = x.size(0)
        seq_len = x.size(1)
        
        outputs_fwd = []
        outputs_bwd = []
        
        hidden_fwd = torch.zeros(batch_size, self.hid_size).to(x.device)
        hidden_bwd = torch.zeros(batch_size, self.hid_size).to(x.device)
        
        for idx in range(seq_len):
            hidden_fwd = torch.tanh(x[:, idx] @ self.w_x_forward + hidden_fwd @ self.w_h_forward + self.b_h_forward)
            hidden_fwd = self.layer_norm_forward(hidden_fwd)
            outputs_fwd.append(hidden_fwd)
            
        for idx in reversed(range(seq_len)):
            hidden_bwd = torch.tanh(x[:, idx] @ self.w_x_backward + hidden_bwd @ self.w_h_backward + self.b_h_backward)
            hidden_bwd = self.layer_norm_backward(hidden_bwd)
            outputs_bwd.append(hidden_bwd)
            
        outputs_fwd = torch.stack(outputs_fwd, dim=1)
        outputs_bwd = torch.stack(outputs_bwd[::-1], dim=1)
        outputs = torch.cat([outputs_fwd, outputs_bwd], dim=-1)
        #print(outputs.shape)
        outputs = F.relu(self.y(outputs))
        return outputs
    
class BidirRNN(nn.Module):
    def __init__(self, vocab_size, emb_size, n_classes, hid_size=64):
        super().__init__()
        self.emb = nn.Embedding(vocab_size, emb_size)
        self.rnn1 = BidirRNNLayer(emb_size, hid_size)
        self.rnn2 = BidirRNNLayer(hid_size * 2, hid_size)
        #self.rnn3 = BidirRNNLayer(hid_size * 2, hid_size)
        #self.rnn4 = BidirRNNLayer(hid_size * 2, hid_size)
        self.out = nn.Linear(hid_size * 2, n_classes)  # Учитываем размер 2 * hid_size из-за bidirectional
    
    def forward(self, x):
        x = x['Text']
        x = self.emb(x)
        
        x = self.rnn1(x)
        x = self.rnn2(x)
        #x = self.rnn3(x)
        #x = self.rnn4(x)
        
        output = F.log_softmax(self.out(x[:, -1, :]), dim=-1)  # Softmax для классификации
        return output
```

```python
N_CLASSES = len(le.classes_)
EMBEDDING_SIZE = 128
HIDDEN_SIZE = 128
BATCH_SIZE = 128
EPOCHS = 5

bidirRNN = BidirRNN(vocab_size, EMBEDDING_SIZE, N_CLASSES, HIDDEN_SIZE).to(device)
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(bidirRNN.parameters(), lr=4e-3)

bidir_rnn_dict = train_loop(bidirRNN, df, df_val, criterion, optimizer, vocab_size, N_CLASSES, 
           EMBEDDING_SIZE, HIDDEN_SIZE, BATCH_SIZE, EPOCHS, device)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class RNNModel2(nn.Module):
    def __init__(self, vocab_size, emb_size, n_classes, hid_size=64, num_layers=2):
        super().__init__()
        self.emb = nn.Embedding(vocab_size, emb_size)
        
        # Добавляем num_layers=2 для двухслойной RNN
        self.rnn1 = nn.RNN(emb_size, hid_size, bidirectional=True, dropout=0.2,num_layers=num_layers, batch_first=True)
        self.out = nn.Linear(hid_size, n_classes)
        
    def forward(self, x):
        x = x['Text']  # Получаем текстовые данные
        x = self.emb(x)  # Векторизация с помощью embedding слоя
        
        # Пропускаем данные через двухслойную RNN
        output, hidden = self.rnn1(x)
        
        # Используем последнее скрытое состояние второго слоя
        # hidden имеет размер (num_layers, batch_size, hid_size), берем последнее состояние второго слоя
        hidden = hidden[-1]  # берем последний слой
        
        # Пропускаем последнее скрытое состояние через fully connected слой
        output = self.out(hidden)
        
        # Применяем log softmax для получения логарифмов вероятностей
        output = F.log_softmax(output, dim=-1)
        
        return output
```

```python
class GRULayer(nn.Module):
    def __init__(self, emb_size, hid_size):
        super().__init__()
        
        self.emb_size = emb_size
        self.hid_size = hid_size
        
        self.w_rx = nn.Parameter(torch.Tensor(emb_size, hid_size))
        self.w_rh = nn.Parameter(torch.Tensor(hid_size, hid_size))
        self.b_r = nn.Parameter(torch.zeros(1, hid_size))

        self.w_zx = nn.Parameter(torch.Tensor(emb_size, hid_size))
        self.w_zh = nn.Parameter(torch.Tensor(hid_size, hid_size))
        self.b_z = nn.Parameter(torch.rand(1, hid_size))

        self.w_nx = nn.Parameter(torch.Tensor(emb_size, hid_size)) 
        self.w_nh = nn.Parameter(torch.Tensor(hid_size, hid_size))
        self.b_n = nn.Parameter(torch.zeros((1, hid_size)))
        
        nn.init.xavier_uniform_(self.w_rx)
        nn.init.xavier_uniform_(self.w_rh)
        
        nn.init.xavier_uniform_(self.w_zx)
        nn.init.xavier_uniform_(self.w_zh)
        
        nn.init.xavier_uniform_(self.w_nx)
        nn.init.xavier_uniform_(self.w_nh)
        
        self.norm = nn.LayerNorm(hid_size)
        self.drop = nn.Dropout(0.1)
    def forward(self, x, hidden = None):
        hidden = torch.zeros((x.size(0), self.hid_size)).to(x.device) if hidden is None else hidden
        out = []
        for idx in range(x.size(1)):
            r = torch.sigmoid(x[:, idx] @ self.w_rx + hidden @ self.w_rh + self.b_r)
            z = torch.sigmoid(x[:, idx] @ self.w_zx + hidden @ self.w_zh + self.b_z)
            n = torch.tanh(x[:, idx] @ self.w_nx + r * (hidden @ self.w_nh + self.b_n))
            hidden = (1 - z) * n + z * hidden
            hidden = self.norm(hidden)
            hidden = self.drop(hidden)
            
            out.append(hidden)
        output = torch.stack(out, dim=1)
        return hidden, output
    
class GRUModel(nn.Module):
    def __init__(self, vocab_size, emb_size, num_classes, hid_size=64):
        super().__init__()
        self.emb = nn.Embedding(vocab_size, emb_size)
        self.gru1 = GRULayer(emb_size, hid_size)
        self.gru2 = GRULayer(hid_size, hid_size)
        self.cls = nn.Linear(hid_size, num_classes)
        self.dropout = nn.Dropout(0.1)

    def forward(self, x):
        x = x['Text']
        x = self.emb(x)
        _, x = self.gru1(x)
        hidden, x = self.gru2(x) 
        hidden = self.dropout(hidden)
        output = self.cls(hidden)
        return output
```

```python
import random
def get_synonym(word):
    """Находит случайный синоним для данного слова, если он существует."""
    synonyms = wordnet.synsets(word)
    if not synonyms:
        return word  # Возвращаем исходное слово, если нет синонимов
    # Берем случайный синоним, извлекая его первую лемму (базовую форму)
    synonym = random.choice(synonyms).lemmas()[0].name()
    return synonym

def augment_text(text, prob=0.3):
    """Возвращает текст с заменой некоторых слов на синонимы."""
    words = text.split()
    new_words = [get_synonym(word) if random.random() < prob else word for word in words]
    return ' '.join(new_words)
```

```python
import torch
from torch.utils.data import Dataset, DataLoader
import random
from nltk.corpus import wordnet
from keras.utils import to_categorical

device = 'cuda' if torch.cuda.is_available() else 'cpu'

class AugmentedData(Dataset):
    
    def __init__(self, texts, labels, n_classes, sent_len, token_to_id, tokenizer, device, train=False, prob=0.5):
        # передаем все: тексты, лейблы, словарь токен ту айди (что бы его создать я использовал копию тренировочного датафрейма)
        # и токенизатор
        self.texts = texts
        self.labels = labels
        self.prob = prob
        self.n_classes = n_classes
        self.sent_len = sent_len 
        self.token_to_id = token_to_id
        self.tokenizer = tokenizer
        self.device = device
        self.train = train
        
        
    def __len__(self):
        return len(self.texts) # база
    
    def __getitem__(self, idx):
        text = self.texts[idx] # получаем текст и его класс
        label = self.labels[idx]
        
        text = text.split(' ') # сплитим текст
        text = [word for word in text if word not in stop_words and len(word) > 0] # удаляем все стоп слова
        text = ' '.join(text)
    
        # теперь можно аугментировать
        if self.train:
            if random.random() < self.prob: # используем аугментацию в соответствии с вероятностью
                text = self.augmentation(text)
        
        label = to_categorical(label, num_classes=self.n_classes) # получаем one hot вектор
        
        encoded = self.tokenizer.encode(text, out_type=str) # токенизируем 
        encoded = ' '.join(encoded) # обратно в строку 
        vector = self.as_vector(encoded) # получаем вектор (аналог семинарской функции as_matrix только для одного предложения а не корпуса)

        
        return self.to_tensors(vector, label) # тензорируем
    
    def get_synonym(self, word): 
        synonyms = wordnet.synsets(word)
        if not synonyms:
            return word  # Возвращаем исходное слово, если нет синонимов
        # Берем случайный синоним, извлекая его первое лемма (базовую форму)
        synonym = random.choice(synonyms).lemmas()[0].name()
        return synonym

    def augmentation(self, text, prob=0.3): 
        words = text.split()
        new_words = [self.get_synonym(word) if random.random() < prob else word for word in words]
        return ' '.join(new_words)
    
    def as_vector(self, sequence):
        """ Convert a single token sequence into a vector with padding """
        UNK_IX, PAD_IX = map(self.token_to_id.get, [UNK, PAD])

        # Разделяем строку на токены
        if isinstance(sequence, str):
            sequence = sequence.split()

        # Создаем вектор нужной длины и заполняем его PAD_IX
        vector = np.full(self.sent_len, PAD_IX, dtype=np.int32)

        # Преобразуем токены в индексы и заполняем вектор
        row_ix = [self.token_to_id.get(word, UNK_IX) for word in sequence[:self.sent_len]]
        vector[:len(row_ix)] = row_ix

        return vector
    
    def to_tensors(self, text, label):
        text_tensor = torch.tensor(text, device=self.device, dtype=torch.int64)
        label_tensor = torch.tensor(label, device=self.device)
        return text_tensor, label_tensor
```

```python
from tqdm.notebook import tqdm
import copy
from sklearn.metrics import f1_score

def calculate_accuracy(output, target):
    _, predicted = torch.max(output, dim=1)
    _, target= torch.max(target, dim=1)
    correct = (predicted == target).sum().item()
    total = target.size(0)
    accuracy = correct / total
    return accuracy
def train_loop(model,
               train_loader,
               val_loader,
               loss_fn, 
               opt,
               epochs,
               device,
               save_best_weights=True,
               print_each = 5,
               scheduler=None):
    

    best_weights = None
    best_f1 = -float('inf')
    acc = loss_score = ctr = 0
    tr_preds = []
    tr_true = []
    metrics = {'train' : {'loss' : [], 'f1': []},
              'val': {'loss' : [], 'f1': []}}
    for epoch in range(epochs):
        print(f'epoch: {epoch+1}')
        model.train()
        for text_batch, label_batch in tqdm(train_loader):
            y_pred = model(text_batch)
            loss = loss_fn(y_pred, label_batch)
            opt.zero_grad()
            loss.backward()
            opt.step()
            
            # считаем и сохраняем метрики
            ctr += 1
            loss_score += loss.item()
            true_labels = torch.argmax(label_batch, dim=1).cpu().numpy()
            pred_labels = torch.argmax(y_pred, dim=1).cpu().numpy()
            tr_preds.extend(pred_labels)
            tr_true.extend(true_labels)
            acc += calculate_accuracy(y_pred, label_batch)
        if scheduler:
            scheduler.step()
        acc /= ctr
        loss_score /= ctr
        f1 = f1_score(tr_true, tr_preds, average='macro')
        if (epoch+1) % print_each == 0:
            print("\nTrain Results:")
            if scheduler:
                print(f'Learning rate: {scheduler.get_last_lr()}')
            print("Accuracy: %.5f" % acc)
            print("Macro F1: %.5f" % f1)
            
            
        metrics['train']['f1'].append(f1)
        metrics['train']['loss'].append(loss_score)
        acc = loss_score = ctr = 0
        tr_true = []
        tr_preds = []
  
        
        model.eval()
        with torch.no_grad():
            val_acc = val_ctr = val_loss = 0
            val_preds = []
            val_true = []
            
            for text_batch, label_batch in val_loader:
                y_pred = model(text_batch)
                loss = loss_fn(y_pred, label_batch)
                
                true_labels = torch.argmax(label_batch, dim=1).cpu().numpy()
                pred_labels = torch.argmax(y_pred, dim=1).cpu().numpy()
                val_preds.extend(pred_labels)
                val_true.extend(true_labels)
                val_acc += calculate_accuracy(y_pred, label_batch)
                val_loss += loss.item()
                val_ctr += 1
        val_acc /= val_ctr 
        val_loss /= val_ctr
        val_f1 = f1_score(val_true, val_preds, average='macro')
        if (epoch+1) % print_each == 0:
            print("\nValid Results:")
            print("Accuracy: %.5f" % val_acc)
            print("Macro F1: %.5f" % val_f1)
        metrics['val']['f1'].append(val_f1)
        metrics['val']['loss'].append(val_loss)
        
        if val_f1 > best_f1:
            best_f1 = val_f1
            best_weights = copy.deepcopy(model.state_dict())
    
    if save_best_weights:
        model.load_state_dict(best_weights)
    return metrics
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class GRUModel2(nn.Module):
    def __init__(self, vocab_size, emb_size, num_classes, hid_size=64, n_layers=2):
        super().__init__()
        self.emb = nn.Embedding(vocab_size, emb_size)
        
        # Добавляем batch_first=True, чтобы входные данные были в формате (batch_size, seq_len, emb_size)
        self.gru1 = nn.GRU(emb_size, hid_size, batch_first=True, num_layers=n_layers, dropout=0.2, bidirectional=True)  
        self.cls = nn.Linear(hid_size * 2, num_classes)

    def forward(self, x):
        x = self.emb(x)  # Преобразуем в эмбеддинги

        # Получаем все скрытые состояния
        states, _ = self.gru1(x)
        
        # Берем последнее скрытое состояние для каждого элемента в батче
        last = states[:, -1, :]  # Используем последнее скрытое состояние
        
        # Пропускаем через полносвязный слой
        output = self.cls(last)
        
        return output
```
